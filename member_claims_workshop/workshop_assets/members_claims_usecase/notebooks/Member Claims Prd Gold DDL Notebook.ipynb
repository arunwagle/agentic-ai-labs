# PRD_GOLD create-tables notebook with PK/FK documentation

This notebook creates the main PRD_GOLD dimensions and facts with:
- **table-level comments**
- **column-level comments**
- **informational primary keys and foreign keys** in Unity Catalog

Important:
- Databricks primary key and foreign key constraints are **informational** and are **not enforced**.
- They are still valuable for documentation, semantic clarity, Genie discoverability, and some query optimizations in Unity Catalog. citeturn474642search0turn474642search1turn474642search4

Relationships implemented where they are clear from the PRD_GOLD ERD:
- `dim_member_history.member_sk -> dim_member.member_sk`
- `dim_member_identifier.member_sk -> dim_member.member_sk`
- `dim_provider.provider_address_sk -> dim_address.address_key`
- `fact_member_enrollment.member_sk -> dim_member.member_sk`
- `fact_claim_header.clm_member_sk -> dim_member.member_sk`
- `fact_claim_header.clm_service_facility_address_sk -> dim_address.address_key`

Not implemented as foreign keys in this notebook because the ERD does not show an unambiguous normalized key relationship:
- role-based provider fields on claims (these are text/NPI/provider-role attributes rather than provider surrogate keys)
- `fact_claim_detail -> fact_claim_header` because the ERD text does not clearly prove whether `clm_claim_id` is the authoritative referenced key vs another surrogate key
